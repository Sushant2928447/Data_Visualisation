```python
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(42)
num_samples = 100
categories = ["A", "B", "C"]
data = {
    "Category": np.random.choice(categories, num_samples),
    "Value": np.random.normal(loc=10, scale=2, size=num_samples)
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)
df
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate x-values from 0 to 10
x_values = np.linspace(0, 10, 100)

# Compute y-values (sine of x)
y_values = np.sin(x_values)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")
p.line(x_values, y_values, line_width=2, legend_label="sin(x)", line_color="blue")

# Show the plot
output_notebook()
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
```

```python
# importing the modules 
from bokeh.plotting import figure, output_file, show 
from bokeh.palettes import magma 
import random 

# file to save the model 
output_file("gfg.html") 
	
# instantiating the figure object 
graph = figure(title = "Bokeh Scatter Graph") 

# name of the x-axis 
graph.xaxis.axis_label = "x-axis"
	
# name of the y-axis 
graph.yaxis.axis_label = "y-axis"

# points to be plotted 
x = [n for n in range(25)] 
y = [random.random() + 1 for n in range(25)] 
size = 10

# color value of the scatter points 
color = magma(25) 

# plotting the graph 
graph.scatter(x, y, 
			size = size, 
			color = color) 
	
# displaying the model 
show(graph) 

```

```python
#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a Bokeh figure
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1), toolbar_location=None)

# Convert data to a flat list for the heatmap
source = ColumnDataSource(data=dict(x=xx.ravel(), y=yy.ravel(), value=data_heatmap.ravel()))

# Add the heatmap glyph
p.rect(x="x", y="y", width=1, height=1, source=source, fill_color="red", line_color=None, alpha=0.7)

# Show the plot
show(p)

```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
"""np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)"""
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame with the provided sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix for the heatmap
sales_heatmap = df.pivot_table(index='Month', columns='Day', values='Sales')

# Create the heatmap
sns.heatmap(sales_heatmap, cmap='YlGnBu', annot=True, fmt='.0f')
plt.xlabel('Day')
plt.ylabel('Month')
plt.title('Sales Variation Heatmap')
plt.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function. z = sin(x**2 +y**2)
"""x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)"""
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Given x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create the 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Customize labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Surface Plot: z = sin(x^2 + y^2)')

# Show the plot
plt.show()
```

```python
# Given dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Calculate bubble size (proportional to population)
bubble_size = df['Population'] * 0.1  

# Create the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=bubble_size, alpha=0.7, c='b', edgecolors='k')

# Customize labels
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: Population vs. GDP')
plt.grid(True)

# Add labels for each country
for i, row in df.iterrows():
    plt.text(row['GDP'], row['Population'], row['Country'], ha='center', va='center')

plt.show()
```
